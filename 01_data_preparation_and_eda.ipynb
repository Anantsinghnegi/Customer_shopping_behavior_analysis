Load customer shopping behavior dataset into a Pandas DataFrame

```python
import pandas as pd
df= pd.read_csv('customer_shopping_behavior.csv')
```

```python
df.head(5) #verifying loaded data
```

```python
df.info()  # checking column data types, and non-null counts
```

```python
df.describe()
```

```python
# Renaming columns according to snake casing for better readability and documentation
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df.isnull().sum() # Checking if missing data or null values are present in the dataset
```

```python
# Filling missing review ratings with category median
category_median = df.groupby('category')['review_rating'].transform('median')
df['review_rating'] = df['review_rating'].fillna(category_median)
```

```python
df.isnull().sum()
```

```python
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# Create age groups to support customer segmentation and data visualization
labels=['Young Adults','Adults','Middle-aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head()
```

```python
(df['discount_applied']==df['promo_code_used']).all()
```

```python
df.drop(columns=['promo_code_used'], inplace=True)
```
